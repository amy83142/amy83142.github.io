#### 型態轉換

"型別轉換會在記憶體空間配置出新的物件，內容為轉換後的結果。"


```python
# 1. 轉換成整數型態 (int)
num_str = "42"
num_int = int(num_str)
print(f"整數型態轉換結果：{num_int}")  # 輸出 42

# 2. 轉換成浮點數型態 (float)
float_str = "3.14"
float_num = float(float_str)
print(f"浮點數型態轉換結果：{float_num}")  # 輸出 3.14

# 3. 轉換成字串型態 (str)
num = 123
num_str = str(num)
print(f"字串型態轉換結果：'{num_str}'")  # 輸出 '123'

# 4. 轉換成列表型態 (list)
str_example = "hello"
list_example = list(str_example)
print(f"列表型態轉換結果：{list_example}")  # 輸出 ['h', 'e', 'l', 'l', 'o']

# 5. 轉換成元組型態 (tuple)
list_example = [1, 2, 3]
tuple_example = tuple(list_example)
print(f"元組型態轉換結果：{tuple_example}")  # 輸出 (1, 2, 3)

# 6. 轉換成布林型態 (bool)
num = 0
bool_example = bool(num)
print(f"布林型態轉換結果：{bool_example}")  # 輸出 False
```

# 字串string(序列型態sequence types) 計算

字串（String）


* 字串的特徵為引號，包含單引號，雙引號和 三引號，再現時預設以單引號表示
*   使用索引位置可以存取元素

*   字串的元素是"有序的"（誰前誰後有關係）
*   字串頭尾的引號必須相同對稱



```python
# 假設有以下的值
x = 3
s = [0, 1, 2, 3, 4]
t = [6, 7, 8]
n = 3
i = 2
j = 5
k = 2
```

```python
## 1. 判斷元素是否在序列中
# 判斷x是否在s中 (x = 3)
result1 = x in s
print(result1)

# 判斷x是否不在s中
result2 = x not in s
print(result2)


## 2. 連接與重複序列
# 連接s與t
result3 = s + t
print(result3)

# 將s重複n次連接s本身 (n = 3)
result4 = n * s
print(result4)


##3. 索引與子序列 (i=2, j=5, k=2)
# 取得索引值i的 "元素"
result5 = s[i]
print(result5)

# 取得索引值i到j的 "子序列"
result6 = s[i:j]
print(result6)

# 取得索引值i到j，間隔k的 "子序列"
result7 = s[i:j:k]   #(由i始到j 每間隔k取1個)
print(result7)


## 4. 回傳序列的相關資訊
# 回傳s的元素個數
result8 = len(s)
print(result8)

# 回傳s中的最小值
result9 = min(s)
print(result9)

# 回傳s中的最大值
result10 = max(s)
print(result10)


## 5. 其他
# 取得s中第一次出現i的"索引值" (第幾個位置)(i=2)
result11 = s.index(i)
print(result11)

# 累計s中i出現的個數 (i=2)
result12 = s.count(i)
print(result12)
```

# 串列(list)

* 是最實用也最常用的群集  
* 可以收集不同資料型態的元素

* list的元素是"有序的"（誰前誰後"有關係"）

```python
list1=[1,2+0j,"three" ,4.0]  # 不同型態的資料可以被收集在同一個清單中
list1[0] #顯示 1  (在電腦領域裡計數由零開始)
list1[1] #顯示 2+0j
list1[2] #顯示 "three"
```

```python
#範例
my_list = [1, 2, 3, 4, 5]
x = 3
i = 2
```

```python
# 將元素 x 附加到列表的最後
my_list.append(x)
print(my_list)  # 輸出 [1, 2, 3, 4, 5, 3]

# 將列表 x 的所有元素附加到列表的最後
my_list.extend([8, 9, 10])
print(my_list)  # 輸出 [1, 2, 3, 4, 5, 3, 8, 9, 10]

# 計算列表中元素 x 的出現次數 (x=3)
count_x = my_list.count(x)
print(count_x)  # 輸出 2

# 回傳元素 x 在列表中的"索引值" (x=3)
index_x = my_list.index(x)
print(index_x)  # 輸出 2

# 將元素 x 插入到列表的索引值 i 的位置 (i=2, x=3)
my_list.insert(i, x)
print(my_list)  # 輸出 [1, 2, "3", 3, 4, 5, 3, 8, 9, 10]

# "移除並回傳"列表中索引值為 i 的元素 (i=2)
removed_element = my_list.pop(i)
print(removed_element)  # 輸出 3 (移除的元素)
print(my_list)  # 輸出 [1, 2, 3, 4, 5, 3, 8, 9, 10]

# 移除列表中"第一個出現的元素x"  (x=3)
my_list.remove(x)
print(my_list)  # 輸出 [1, 2, 4, 5, 3, 8, 9, 10]

# "倒轉"列表中元素的順序
my_list.reverse()
print(my_list)  # 輸出 [10, 9, 8, 3, 5, 4, 2, 1]

# 對列表中的元素進行排序
my_list.sort()
print(my_list)  # 輸出 [1, 2, 3, 4, 5, 8, 9, 10]
```

# 上午作業#計算BMI

######成人的體重分級與標準


分   級      身體質量指數
  



體重過輕    BMI ＜ 18.5

正常範圍    18.5 ≦ BMI ＜24

過   重    24 ≦ BMI ＜ 27

輕度肥胖    27 ≦ BMI ＜ 30

中度肥胖    30 ≦ BMI ＜ 35

重度肥胖    BMI ≧ 35

```python
#計算BMI
num1=float(input('輸入身高(cm):'))/100 #公尺
num2=float(input('輸入體重(kg):'))
bmi=num2/(num1*num1)
print(f"您的BMI:{bmi}")    #"f"為格式化文字用

if bmi<18.5:
  print('分級結果:體重過輕')
elif 18.5<=bmi<24:
  print('分級結果:正常範圍')
elif 24<=bmi<27:
  print('分級結果:過重')
elif 27<=bmi<30:
  print('分級結果:輕度肥胖')
elif 30<=bmi<35:
  print('分級結果:中度肥胖')
elif 35<=bmi:
  print('分級結果:重度肥胖')
```

## 補值

```python
# 1.平均值補值（Mean Imputation）：使用該欄位的平均值填補缺失值
#'HR', 'NBPs', 'NBPd', 'T', 'RR', 'SpO2'

df[['HR']] = df[['HR']].fillna(df[['HR']].mean())
```

```python
# 2.中位數補值（Median Imputation）：使用該欄位的中位數填補缺失值。
df[['NBPs', 'NBPd']] = df[['NBPs', 'NBPd' ]].fillna(df[['NBPs', 'NBPd']].median())
```

```python
# 3.前後值補值（Forward Fill / Backward Fill）：使用缺失值前面或後面的值填補。
df[['T']] = df[['T']].fillna(method='ffill')  # Forward Fill
# 或
df[['RR']] = df[['RR']].fillna(method='bfill')  # Backward Fill
```

```python
# 4.插值法補值（Interpolation）：使用相鄰值的插值填補。
df[['SpO2']] = df[['SpO2']].interpolate(method='linear')
```
